```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=2000, n_features=2, centers=4, cluster_std=1.6, random_state=101)
```

```python
data[0]
```

```python
print(data[0].shape, data[1].shape)
```

```python
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4, max_iter=400)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,6), sharey=True)

ax[0].set_title('Original')
ax[1].set_title('KMeans')

sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=data[1], ax=ax[0])
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=km.labels_, ax=ax[1])
```
